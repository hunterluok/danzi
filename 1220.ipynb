```python
from pandas import read_excel
import pandas as pd
```

```python
data = pd.read_excel("C:/Users/Administrator/Desktop/ziji/1220/data.xlsx",sheetname='sheet1')
print(data.head())
```

```python
def g(k):
    name = []
    for i in range(1,k+1):
        na = "Q"+str(i)
        name.append(na)
    return name
print(g(11))

```

```python
def f(j):
    t = []
    for i in data.columns:
        if i.split('.')[0]==str(j):
            t.append(i)
            d = data[t]
    return d
#print(f(2))
```

```python
tx1 = f(1)
tx1.columns=g(11)
tx1['类别']=1
#print(tx1)
tx2 = f(2)
tx2.columns=g(11)
tx2['类别']=2

tx3 = f(3)
tx3.columns=g(11)
tx3['类别']=3

tx4 = f(4)
tx4.columns=g(11)
tx4['类别']=4

tx5 = f(5)
tx5.columns=g(11)
tx5['类别']=5

tx6 = f(6)
tx6.columns=g(11)
tx6['类别']=6

tx7 = f(7)
tx7.columns=g(11)
tx7['类别']=7

tx8 = f(8)
tx8.columns=g(11)
tx8['类别']=8

tx9 = f(9)
tx9.columns=g(11)
tx9['类别']=9

tx10 = f(10)
tx10.columns=g(11)
tx10['类别']=10

tx11 = f(11)
tx11.columns=g(11)
tx11['类别']=11

tx12 = f(12)
tx12.columns=g(11)
tx12['类别']=12

tx13 = f(13)
tx13.columns=g(11)
tx13['类别']=13

tx14 = f(14)
tx14.columns=g(11)
tx14['类别']=14

tx15 = f(15)
tx15.columns=g(11)
tx15['类别']=15

tx16 = f(16)
tx16.columns=g(11)
tx16['类别']=16
```

```python
nd =pd.concat([tx1,tx2,tx3,tx4,tx5,tx6,tx7,tx8,tx9,tx10,tx11,tx12,tx13,tx14,tx15,tx16],axis=0)
print(nd.head())
```

```python
nd['涉入度']= round(nd[['Q1','Q2','Q3','Q4','Q5']].mean(axis=1),2)
nd['用户体验']= round(nd[['Q6','Q7','Q8']].mean(axis=1),2)
nd['用户态度']= round(nd[['Q9','Q10','Q11']].mean(axis=1),2)
print(nd.head())
```

```python
for i in nd['类别']:
    if i < 9:
        nd.loc[nd['类别']==i,'营销']= 1
    else:
        nd.loc[nd['类别']==i,'营销']= 2
# print(nd.loc[nd['类别']==9,])

for i in nd['类别']:
    if i % 2 ==1:
        nd.loc[nd['类别']==i,'位置']= 1
    else:
        nd.loc[nd['类别']==i,'位置']= 2
```

```python
for i in nd['类别']:
    if i <5:
        nd.loc[nd['类别']==i,'引导']= 1
    elif i>4 and i<9:
        nd.loc[nd['类别']==i,'引导']= 2
    elif i>8 and i<13:
        nd.loc[nd['类别']==i,'引导']= 1         
    else:
        nd.loc[nd['类别']==i,'引导']= 2
#print(nd)
    
```

```python
for i in nd['类别']:
    if i in [1,2,5,6,9,10,13,14]:
        nd.loc[nd['类别']==i,'数量']= 1
    else:
        nd.loc[nd['类别']==i,'数量']= 2
```
